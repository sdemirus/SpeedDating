### 2. Data Loading and Exploration

This code snippet will load the dataset and display its shape, column names, data types, and summary statistics to get an initial understanding of the data.

````python
# Importing necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
data = pd.read_csv('Data/speed_data.csv')
````

###### 1. Context of the Dataset:

This dataset comes from a Speed Dating experiment conducted by Columbia Business School professors. The study's aim was to investigate what makes people attracted to one another in a speed dating setting. Participants in the study went on a series of four-minute dates with other participants and then filled out a survey after each date.

The numerical scores in the dataset represent the participants' ratings on various attributes such as attractiveness, sincerity, intelligence, and fun, among others. These scores were given on a scale from 1 to 10. The participants also answered questions about their own attributes and preferences.

Here is a breakdown of how some of the variables were measured:
- Attractiveness, Sincerity, Intelligence, Fun: Rated on a scale from 1 to 10 by the participant's dates.
- Age: Self-reported by participants.
- Gender: Self-reported by participants (1 for female, 0 for male).
- Career field: Self-reported by participants and encoded numerically in the dataset.
- Goals: Participants' responses to why they attended the event, encoded numerically in the dataset.
"""


````python
data.head()
````

````python
# Display basic information about the dataset
print("Shape of the dataset:", data.shape)
print("\nColumns in the dataset:", data.columns)
print("\nData types of columns:\n", data.dtypes)
print("\nSummary statistics:\n", data.describe())
````

##### 2. Supporting Insights with Data Points

````python
# Age and Gender Distribution Analysis

age_distribution = data['age'].dropna()
gender_distribution = data['gender'].value_counts(normalize = True)
````

````python
# Distribution of age
plt.figure(figsize=(10, 6))
sns.histplot(data['age'], bins=20, kde=True, color='skyblue')
plt.title('Distribution of Age')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
````

````python
# Gender distribution
plt.figure(figsize=(8, 5))
data['gender'].value_counts().plot(kind='bar', color=['skyblue', 'pink'])
plt.title('Gender Distribution')
plt.xlabel('Gender')
plt.ylabel('Count')
plt.xticks(rotation=0)
plt.show()
````

````python
# Insights with supporting data points
mean_age = age_distribution.mean()
median_age = age_distribution.median()
female_percentage = gender_distribution[1] * 100
male_percentage = gender_distribution[0] * 100

print(f"The mean age of participants is {mean_age:} years.")
print(f"The median age of participants is {median_age:} years.")
print(f"Percentage of female participants: {female_percentage:}%")
print(f"Percentage of male participants: {male_percentage:}%")
````

````python
# Interests and Goals Analysis
career_distribution = data['career'].value_counts(normalize=True) * 100
````

````python
# Interests and Goals Analysis
# Distribution of goals
plt.figure(figsize=(10, 6))
data['goal'].value_counts().plot(kind='bar', color='skyblue')
plt.title('Distribution of Goals')
plt.xlabel('Goal')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()
````

````python
# Let's consider the top 5 careers for clarity
top_careers = career_distribution.head(5)
print("Top 5 careers among participants (in percentage):")
print(top_careers)
````

````python
# Distribution of career
plt.figure(figsize=(12, 6))
sns.countplot(data=data, x='career', order=data['career'].value_counts().index[:10], palette='viridis')
plt.title('Top 10 Careers of Participants')
plt.xlabel('Career')
plt.ylabel('Count')
plt.xticks(rotation=45, ha='right')
plt.show()
````

````python
# Rating Attributes Analysis
attribute_means = data[['attr', 'sinc', 'intel', 'fun']].mean()

print("Mean ratings for attributes:")
print(attribute_means)
````

````python
# Attraction Ratings Analysis
# Distribution of ratings
ratings_cols = ['dec', 'attr', 'sinc', 'intel', 'fun', 'amb', 'shar']
ratings_data = data[ratings_cols]

plt.figure(figsize=(12, 8))
sns.boxplot(data=ratings_data, palette='Set2')
plt.title('Distribution of Ratings')
plt.ylabel('Rating')
plt.xticks(rotation=45)
plt.show()
````

````python
# Insight about higher ratings for attractiveness and sincerity
higher_ratings = attribute_means[['attr', 'sinc']].mean()
print(f"Average rating for attractiveness and sincerity: {higher_ratings: }")
````

##### 3. Graph Analysis

````python
# Removing 'dec' variable from Attraction Ratings Analysis
attributes = ['attr', 'sinc', 'intel', 'fun']
plt.figure(figsize=(10, 6))
sns.boxplot(data=data[attributes])
plt.title('Distribution of Importance Ratings for Various Attributes')
plt.show()
````

````python
# Decoding the goals for Distribution of Goals graph
goal_labels = {1: 'Seemed like a fun night out', 2: 'To meet new people', 3: 'To get a date', 4: 'Looking for a serious relationship'}
data['goal'] = data['goal'].map(goal_labels)
````

````python
plt.figure(figsize=(10, 6))
sns.countplot(data=data, x='goal')
plt.title('Distribution of Goals for Attending the Event')
plt.xticks(rotation=45)
plt.show()
````

##### 4. Further Analysis Opportunities

````python
data.head()
````

````python
# Decision Making Analysis
# Exploring average distributions by decision
decision_gender_distribution = data.groupby('dec')['gender'].mean()
decision_age_distribution = data.groupby('dec')['age'].mean()
decision_rating_means = data.groupby('dec')[['attr', 'sinc', 'intel', 'fun']].mean()
````

````python
print("Average gender distribution by decision (1 = female, 0 = male):")
print(decision_gender_distribution)

print("Average age distribution by decision:")
print(decision_age_distribution)

print("Average attribute ratings by decision:")
print(decision_rating_means)
````

````python
# Decision Making Analysis
# Distribution of decisions
plt.figure(figsize=(8, 5))
data['dec'].value_counts().plot(kind='bar', color=['skyblue', 'pink'])
plt.title('Distribution of Decisions')
plt.xlabel('Decision')
plt.ylabel('Count')
plt.xticks(rotation=0)
plt.show()
````
